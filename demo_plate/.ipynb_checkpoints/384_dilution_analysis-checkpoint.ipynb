```python
import pandas as pd
import numpy as np
import sys
sys.path.append('../code')
import serology_analysis_functions as sf
```

```python
fname='demo_serology_plate.xlsx'
sname='demo_data'
nrows=162
df=pd.read_excel(fname,sname,usecols="D:H",skiprows=0,nrows=nrows)
df=df.reset_index(drop=True)
df
```

```python
labels=pd.read_excel(fname,sname,usecols="B:B",nrows=nrows)
#labels2=pd.read_excel(fname,sname,usecols="A:A",nrows=nrows)
labels
```

```python
bsadf=pd.read_excel(fname,sname,usecols="C:C",nrows=nrows)
bsadf=bsadf.reset_index(drop=True)
bsadf
```

```python
#concentrations are halved with each dilution from left to right
concs=[1.0/(4**i) for i in range(0,5)]
```

```python
concs
```

```python
#now parse out the columns--first two are BSA then alternate RBD, SPIKE, ...
dfspk1=df.T
colnames=list(labels.iloc[:,0].values)
for i in range(len(colnames)):
    colnames[i]=colnames[i]+'_'+str(i)
dfspk1.columns=colnames
dfspk1.columns.name=''
dfspk1['conc']=concs
colnames=['conc']+colnames
dfspk1=dfspk1[colnames]
dfspk1=dfspk1.reset_index(drop=True)
dfspk1=dfspk1.astype(float)
dfspk1
```

```python
outdf,allprofiles,allfits=sf.fitDF2(dfspk1,3)
```

```python
outdf['BSA']=bsadf['BSA 1:100']
outdf
```

```python
print('AUC Plot:')
sf.makeHoverAUCPlot(outdf)
```

```python
halfcurves=int(len(labels)/2)
sf.plotCurves(allprofiles,allfits,outdf,dfspk1['conc'].values,'WellName',ymax=3,cols=halfcurves,logplot=True)
```

```python
#output the data
outdf.to_csv('serology_analysis_results1.csv')
dfspk1.to_csv('serology_traces1.csv')
```
